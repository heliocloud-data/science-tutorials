# <font color="blue">Steps</font>

## *<font color="Red">1. Create your own directory in the shared environment: </font>*

You should enter in your "home" directory.  To create your shared directory on the cloud (and have more than 10G of space), you need to create a new directory in the /efs directory.  First, navigate to the /efs folder (gray folder directly under the search window at the upper left). 

Once you're in /efs,  click the icon at the upper left that is a folder with a + sign, then give it your user name.

Refresh the window and your name should now appear in the directory list.



<br>

<br>

## *<font color="Red" style ="border:2px; border-style: dashed; border-color:#000000; padding: .3em;">Accessing the Terminal</font>*


<p style="border:3px; border-style:solid; border-color:#FF0000; padding: 1em;">To access the terminal, click on the blue + at the top left. If the you don't see it then click on the first folder (it has the yellow box around it in the image above), this will display the file browser options and now you should be able to see it.  <br> This will open a new launcher, At the bottom under other select terminal.</p>

<br>

## *<font color="Red">2. Make sure youre at home (just in case you were doing something before that)</font>*
    cd  ~

<br>

## *<font color="Red">3.Make your environment.yml file.  Either add an environment.yml file to your home directory, or make a copy of the one from the /Tutorials/efs  and put it in your home directory. </font>*
* cp efs/Tutorials/test_environment.yml ~/environment.yml

  
  
   

<br>

## *<font color="Red">4. Edit the environment.yml file by clicking on the environment file name at left <br><br>MAKE SURE YOU ARE EDITING THE ONE IN YOUR DIRECTORY AND NOT TUTORIALS!  The directory name at left should say /efs/Yourname and not /efs/Tutorials.  <br><br>Change the name of the project, and add any python packages you will need (the common ones are already listed). </font>*  


<br>

## *<font color="Red">5. Create the conda environment</font>*

Go to the Terminal tab that you created and type the text below. 

    conda env create -f environment.yml


2021-06-29 (1) step4.png

<br>

## *<font color="Red">6. Then activate the environment.  Use the name of the environment at the top of the yml file after name: </font>*

    conda activate your-project

2021-06-29 (3) step 5.png


<br>

## *<font color="Red">7. Check to see if the python environment is set up properly. Type the following: </font>*

    which python

 * It should return /srv/conda/envs/notebook/bin/python or something like that.

<br>

## *<font color="Red">8. If successful, restart the server so it properly resets the environment. </font>*
