# SFTP Information

SFTP is enabled for selected S3 buckets. We will use the helio-dh-data bucket which is backed by the SFTP site 'data.hsdcloud.org' service. Other S3 buckets and EFS storage may have sftp service available. Check with the adminsitrator of your cloud environment.

To make SFTP work, you should have both a public and private RSA key for your account. These are referred to in the tutorial below as "<username_rsa_private>" and "<username_rsa_public>" below.

## Linux Use


1. Copy public and private keys to ~/.ssh folder on your machine

2. chmod 600 username_rsa_public username_rsa_private

3. sftp -i username_rsa_private username@data.hsdcloud.org

## Windows Use

There are 2 known ways to connect to the SFTP service

### 1. Command Prompt

In a CMD window, do a “where sftp” to check that sftp is installed. This should looks like this:

<p>
C:\Users\username>where sftp
C:\Windows\System32\OpenSSH\sftp.exe
</p>
 
With verified, go to the folder with the keys and use the command:

<p>sftp -i aanarock_rsa.dat aanarock@data.hsdcloud.org<p>
 

### 2. PuTTY

1. First, check that the Putty suite installed and in your path.
